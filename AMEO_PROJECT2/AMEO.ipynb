### Summary of dataset


The dataset contains the employment outcomes of engineering graduates as dependent variables (Salary, Job Titles, and Job Locations) along with the standardized scores from three different areas – cognitive skills, technical skills and personality skills. The dataset also contains demographic features. The dataset  contains  around  40 independent variables and 4000 data points. The independent variables are both continuous and categorical in nature. The dataset contains a unique identifier for each candidate. Below mentioned table contains the details for the original dataset.  


**ID** -------------------- A unique ID to identify a candidate <br>
**Salary** -----------------Annual CTC oﬀered to the candidate (in INR)<br>
**DOJ** --------------------Date of joining the company<br>
**DOL** --------------------Date of leaving the company<br>
**Designation** ------------Designation oﬀered in the job<br>
**JobCity** -------------- Location of the job (city)<br>
**Gender** -------------- Candidate’s gender<br>
**DOB**	------------------Date of birth of candidate<br>
**10percentage** ----------Overall marks obtained in grade 10 examinations<br>
**10board**	-------------- The school board whose curriculum the candidate followed in grade 10<br>
**12graduation** ----------- year of graduation - senior year high school<br>
**12percentage** ------------Overall marks obtained in grade 12 examinations<br>
**12board** --------------The school board whose curriculum the candidate followed in grade 12<br>
**CollegeID** -------------Unique ID identifying the college which the candidate attended<br>
**CollegeTier** ------------Tier of college<br>
**Degree** --------------Degree obtained/pursued by the candidate<br>
**Specialization** -------------Specialization pursued by the candidate<br>
**CollegeGPA** --------------Aggregate GPA at graduation<br>
**CollegeCityID** -------------A unique ID to identify the city in which the college is located in<br>
**CollegeCityTier** -----------The tier of the city in which the college is located<br>
**CollegeState** -------------Name of States<br>
**GraduationYear** -----------Year of graduation (Bachelor’s degree)<br>
**English** -----------------Scores in AMCAT English section<br>
**Logical** -----------------Scores in AMCAT Logical section<br>
**Quant** -------------------Scores in AMCAT Quantitative section<br>
**Domain** -----------------Scores in AMCAT’s domain module<br>
**ComputerProgramming** ----Score in AMCAT’s Computer programming section<br>
**ElectronicsAndSemicon** ---Score in AMCAT’s Electronics & Semiconductor Engineering section<br>
**ComputerScience** -----------Score in AMCAT’s Computer Science section<br>
**MechanicalEng** ------------Score in AMCAT’s Mechanical Engineering section<br>
**ElectricalEng** -----------Score in AMCAT’s Electrical Engineering section<br>
**TelecomEng** --------------Score in AMCAT’s Telecommunication Engineering section<br>
**CivilEng** ----------------Score in AMCAT’s Civil Engineering section<br>
**conscientiousness** --------Scores in one of the sections of AMCAT’s personality test<br>
**agreeableness** ------------Scores in one of the sections of AMCAT’s personality test<br>
**extraversion** -----------Scores in one of the sections of AMCAT’s personality test<br>
**neuroticism** ------------Scores in one of the sections of AMCAT’s personality test<br>
**openess_to_experience** --- Scores in one of the sections of AMCAT’s personality test<br>


### Objective
Our goal in conducting Exploratory Data Analysis on the AMEO dataset is to understand the employment outcome for engineering graduates 


```python
#import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Data Cleaning
df.drop(columns = ['Unnamed: 0'], inplace = True)
df['DOJ'] = pd.to_datetime(df['DOJ'])
df['DOB'] = pd.to_datetime(df['DOB'])
```

```python
# Grouping categorical and nummerical columns
cat_columns = []
num_columns = []
for i in df.columns:
    if df[i].dtype == 'object':
        cat_columns.append(i)
    else:
        num_columns.append(i)
```

#### Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” Test this claim with the data given to you


```python
df[df['Specialization'] == ('computer science & engineering' or 'computer engineering')]['Designation'].str.contains('programming Analyst' or 'software Engineer' or 'Hardware Engineer'or 'Associate Engineer').value_counts()
#No computer science engineering graduate is a programming analyst, software engineer, hardware engineer or associate engineer
```

```python
df['Specialization'].value_counts() 
```

```python
df[df['Specialization'].str.contains('computer')]['Designation'].str.contains('programmer analyst' or 'software engineer' or 'hardware engineer' or 'associate engineer').value_counts()
```

```python
#engineers with a computer related degree and are either Programming Analyst, Software Engineer, Hardware Engineer or Associate Engineer
mask1 = df['Specialization'].str.contains('computer')
mask2 = df['Designation'].str.contains('programmer analyst' or 'software engineer' or 'hardware engineer' or 'associate engineer')
masked_df = df[mask1 & mask2]
len(masked_df)
```

```python
len(masked_df['Salary'] >= 250000)
```

```python
#After masking, lets find out what they earn
masked_df['Salary']
```

```python
masked_df['Salary'].describe()
#it can be seen that even 25% of these engineers earn over 3 lakhs
```

```python
#Analysing for fresh graduates
mask3 = df['GraduationYear'] >= 2015
masked_df2 = df[mask1 & mask2 & mask3]
len(masked_df2)
```

```python
masked_df2['Salary']
#so the claim by times of india article was true
```
